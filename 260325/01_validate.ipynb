# 붓꽃 종류 분류하기

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
from sklearn.datasets import load_iris
import pandas as pd
import os
import pickle
import matplotlib.pyplot as plt
```

## 데이터 확인

각각의 속성은 cm 단위로 측정한 값
- 꽃받침 길이(sepal length)
- 너비(sepal width)
- 꽃잎 길이(petal length)
- 너비(petal width)

```python
iris = load_iris()
iris.data[:5]
```

## 데이터 프레임으로 변환

```python
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df["target"] = pd.Series(iris.target)
iris_df.head()
```

## describe

```python
iris_df.describe()
```

## 라벨 확인

```python
iris_df["target"].value_counts()
```

## 데이터, 라벨 준비

```python
x = iris_df.iloc[:, :4]
y = iris_df.iloc[:, -1]
```

```python
def iris_knn(x, y, k):
    train_data, test_data, train_label, test_label = train_test_split(x, y, test_size=0.3)
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(train_data, train_label)
    y_pred = knn.predict(test_data)
    return knn, metrics.accuracy_score(test_label, y_pred), y_pred


for k in range(1,31,5):
    model, scores, y_pred_all = iris_knn(x, y, k)
    print(f"n_neighbors가 {k}일 때 정확도 : {scores:.3f}")
```

## 결과 시각화

붉을 수록 결과가 많은 것.  
그래프의 왼쪽 아래부터 오른쪽 위 사선으로 1,2,3 클래스  

conf_mat을 보면 잘못 검사한 결과도 확인 가능 함.

```python
k = 16
knn, scores, y_pred_all = iris_knn(x, y, k)
y_pred_all = knn.predict(iris.data)
print(f"n_neighbors가 {k}일 때 정확도 : {scores:.3f}")

conf_mat = metrics.confusion_matrix(iris.target, y_pred_all)
print(conf_mat)

# 둘 다 같은 그래프. 색만 다름
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.hist2d(iris.target, y_pred_all, bins=(3,3), cmap=plt.cm.jet)
plt.subplot(1,2,2)
plt.hist2d(iris.target, y_pred_all, bins=(3,3), cmap=plt.cm.binary)

plt.tight_layout()
plt.show()
```

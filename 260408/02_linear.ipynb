# Scikit-Learn을 이용한 선형 회귀

## 1. 패키지 설정

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn import linear_model
```

## 2. 데이터 준비

```python
DATA_URL = "https://github.com/dknife/ML/raw/main/data/"
lin_data = pd.read_csv(DATA_URL + "pollution.csv")
print(lin_data.head())
```

## 3. 데이터 전처리

```python
x = lin_data['input'].to_numpy()
y = lin_data['pollution'].to_numpy()
x = x[:, np.newaxis]  # x를 2차원 배열로 변환하여 선형회귀 모델에 입력할 수 있도록 함
print(x[:5])
```

## 4. 기계 학습 진행

```python
regr = linear_model.LinearRegression()
regr.fit(x, y)
```

```python
lin_data.plot(x="input", y="pollution", kind="scatter")
y_pred = regr.predict([[0], [1]])
plt.plot([0, 1], y_pred)
plt.show()
```

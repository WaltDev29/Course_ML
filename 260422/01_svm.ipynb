## 1. 패키지 설정

```python
from sklearn import svm
import numpy as np
import matplotlib.pyplot as plt
```

## 2. 데이터 준비

```python
# 학습용 데이터 (기존 개체)
X_train = np.array([[2, 3], [1, 2], [1,1]])  # 입력
y_train = np.array([1, -1, -1])           # 라벨
```

```python
# 테스트 데이터
X_test = np.array([[3,3]])  # 입력
```

## 3. 탐색적 데이터 분석

```python
# 학습용 데이터 산포도
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train)

# 테스트용 데이터 산포도
plt.scatter(X_test[:, 0], X_test[:, 1], c='red', marker = 'D', s=100)
plt.xlabel('x1')
plt.ylabel('x2')
plt.xlim(0, 5)
plt.ylim(0, 5)
plt.show()
```

## 4. 모형화 및 학습

```python
# SVM 분류 모형화
clf = svm.SVC(kernel='linear')
```

```python
# 모형 학습
clf.fit(X_train, y_train)
```

## 5. 학습 결과 확인

```python
print("분류별 서포트 벡터의 수: ", clf.n_support_)     # 분류별 서포트 벡터의 수
print("서포트 벡터:", clf.support_vectors_)    # 서포트 벡터
print("서포트 벡터의 색인:", clf.support_)      # 서포트 벡터의 색인 (x_train에서의 인덱스 번호)
print("클래스:", clf.classes_)            # 각 개체들의 클래스
print("가중치:", clf.coef_)               # W: 가중치
print("편향:", clf.intercept_)          # b: 편향
```

## 6. 예측

```python
# 테스트 데이터의 분류
y_pred = clf.predict(X_test)
print(y_pred)
```

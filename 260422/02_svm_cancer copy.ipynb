# 유방암 데이터 SVM으로 분류

## 1. 패키지 설정

```python
from sklearn.datasets import load_breast_cancer
from sklearn import svm
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
import seaborn as sns
```

## 2. 데이터 준비

```python
# 데이터 불러오기
data = load_breast_cancer(as_frame=True)
```

## 3. 탐색적 데이터 분석

```python
print(data.DESCR)
```

```python
print(data.target_names)     # 목표 값 출력
```

```python
# sns.pairplot(data.frame, hue='target')
```

## 4. 데이터 분리

```python
# selected_columns = data.data.iloc[:, -12:]  # 데이터 프레임에서 12개 열만 선택
```

```python
# 학습 데이터와 테스트 데이터를 7:3으로 분리
X_train, X_test, y_train, y_test = train_test_split(data.data, data.target, test_size=0.3, random_state=1234)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

## 5. 피처 스케일링

```python
# 피처 스케일링: 학습 데이터
scalerX = StandardScaler()
scalerX.fit(X_train)
X_train_std = scalerX.transform(X_train)
print(X_train_std)
```

```python
# 피처 스케일링: 테스트 데이터
X_test_std = scalerX.transform(X_test)
print(X_test_std)
```

## 6. 모형화 및 학습

```python
# SVM 분류 모형화: 선형분리
clf = svm.SVC(kernel='linear')
```

```python
# 모형 학습
clf.fit(X_train_std, y_train)
```

## 7. 예측 및 평가

```python
# 테스트 데이터의 분류
y_pred = clf.predict(X_test_std)
print(y_pred)
```

```python
# confusion matrix (혼동 행렬) 확인
cf = confusion_matrix(y_test, y_pred)
print(cf)
```

```python
# 테스트 데이터에 대한 정확도
clf.score(X_test_std, y_test)
```

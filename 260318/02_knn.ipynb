# k-NN 알고리즘 
k-Nearest Neighbor
- k값은 개발자가 직접 설정  
- 정확도가 높은 k값 중 가장 낮은 값을 사용 (연산량을 최소화하기 위해)

## 그래프로 데이터 확인

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.neighbors import KNeighborsClassifier

# 닥스훈트 길이 & 높이
dach_lengh = [77, 78, 85, 83, 73, 77, 73, 80]
dach_height = [25, 28, 29, 30, 21, 22, 17, 35]

# 사모예드 길이 & 높이
samo_lengh = [75, 77, 86, 86, 79, 83, 83, 88]
samo_height = [56, 57, 50, 53, 60, 53, 49, 61]

newdata_lengh = [79]
newdata_height = [35]

# 점 그래프로 표현
plt.scatter(dach_lengh, dach_height, c="red", label="Dachshund")
plt.scatter(samo_lengh, samo_height, c="blue", marker="^", label="Samoyed")
plt.scatter(newdata_lengh, newdata_height, s=100, c="green", marker="p", label="new Data")

plt.xlabel("Length")
plt.ylabel("Height")
plt.title("Dog size")
plt.legend(loc="upper left")

plt.show()
```

## 데이터 전처리

### 라벨링

```python
d_data = np.column_stack((dach_lengh, dach_height))
d_label = np.zeros(len(d_data)) # 0으로 라벨링

s_data = np.column_stack((samo_lengh, samo_height))
s_label = np.ones(len(s_data))  # 1로 라벨링

newdata = [[82, 40]]
```

### 데이터, 라벨링 병합

```python
dogs = np.concatenate((d_data, s_data))
labels = np.concatenate((d_label, s_label))
```

## k-NN 알고리즘 실행

```python
dog_classes = {0:"Dachshund", 1:"Samoyed"}

k = 3
knn = KNeighborsClassifier(n_neighbors=k)

# 학습 진행
knn.fit(dogs, labels)

# 새로운 결과에 대해 예측결과 반환
y_pred = knn.predict(newdata)

print(f"데이터 : {newdata}\n판정결과 : {dog_classes[y_pred[0]]}")
```

## $80 * 사람 수 + $10 * 맥주 개수 + $40

**f(x,y,z) = x + y + z**  
티켓비용 + 맥주 비용 + 주차비(현재는 1~10명만 계산하기 때문에 상수)  
x축을 사람 수, y축을 맥주 개수, z축을 총 비용으로 표현 (3차원)

```python
number_people = np.arange(1, 11)
number_beer = np.arange(0, 20)
number_people, number_beer = np.meshgrid(number_people, number_beer)
print("======== number_people ========\n", number_people)
print("======== number_beer ========\n",number_beer)
total_cost = 80 * number_people + 10 * number_beer + 40

# from mpl_toolkits.mplot3d import Axes3D

fig, axes = plt.subplots(2, 3,
    subplot_kw={'projection':'3d'},
    figsize=(9,6),
    tight_layout=True
)

angles = [0,45,90,135,180]

for ax, angle in zip(axes.flat, angles):
    ax.plot_surface(number_people, number_beer, total_cost)
    ax.set_xlabel("People")
    ax.set_ylabel("Beers")
    ax.set_zlabel("TotalCost")
    ax.azim = angle

axes.flat[-1].axis('off')
fig.tight_layout()
plt.show()
```

## $80 * 사람 수 + $10 * 맥주 개수 + $5 * 핫도그 개수 + $40

```python
def np_cartesian_product(*arrays):
    ''' some numpy kung-fu to produce all
        possible combinations of input arrays '''
    ndim = len(arrays)
    return np.stack(np.meshgrid(*arrays), axis=-1).reshape(-1, ndim)



number_people = np.array([2, 3])
number_beers = np.array([0, 1, 2])
number_hotdogs = np.array([2, 4])
costs = np.array([80, 10, 5])
columns = ["People", "Beer", "HotDogs", "TotalCost"]
counts = np_cartesian_product(number_people,number_beers,number_hotdogs)
totals = (costs[0] * counts[:, 0] +
costs[1] * counts[:, 1] +
costs[2] * counts[:, 2] + 40)
print(pd.DataFrame(np.c_[counts, totals], columns=columns).head(8))
```

```python
number_people = np.array([2, 3])
number_beers = np.array([0, 1, 2])
number_hotdogs = np.array([2, 4])
costs = np.array([80, 10, 5])
columns = ["People", "Beer", "HotDogs", "TotalCost"]
counts = np_cartesian_product(number_people,
number_beers,
number_hotdogs)
totals = np.dot(counts, costs) + 40
print(pd.DataFrame(np.column_stack([counts, totals]), columns=columns).head(8))
```

# 붓꽃 종류 분류하기

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
from sklearn.datasets import load_iris
import pandas as pd
import os
import pickle
```

## 데이터 확인

각각의 속성은 cm 단위로 측정한 값
- 꽃받침 길이(sepal length)
- 너비(sepal width)
- 꽃잎 길이(petal length)
- 너비(petal width)

```python
iris = load_iris()
iris.data[:5]
```

## 데이터 프레임으로 변환

```python
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df["target"] = pd.Series(iris.target)
iris_df.head()
```

## describe

```python
iris_df.describe()
```

## 라벨 확인

```python
iris_df["target"].value_counts()
```

## 데이터, 라벨 준비

```python
x = iris_df.iloc[:, :4]
y = iris_df.iloc[:, -1]
```

```python
def iris_knn(x, y, k):
    train_data, test_data, train_label, test_label = train_test_split(x, y, test_size=0.3)
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(train_data, train_label)
    y_pred = knn.predict(test_data)
    return knn, metrics.accuracy_score(test_label, y_pred)

for k in range(1,31,5):
    model, scores = iris_knn(x, y, k)
    print(f"n_neighbors가 {k}일 때 정확도 : {scores:.3f}")
```

## 모델 저장

```python
os.mkdir("model")
with open("model/iris_model.pkl", "wb") as f:
    pickle.dump(model, f)

print("모델 저장 완료")
```

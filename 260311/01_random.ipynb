# 적은 샘플로 발생하는 오해/선입견

데이터가 적으면 선입견이 발생할 수 있음.  
주사위를 10번 돌렸을 때와 1000번 돌렸을 때의 결과로 실습. 

```python
import numpy as np
import matplotlib.pyplot as plt
```

## 주사위 굴려보기

```python
few = np.random.randint(1,7, size=10)
many = np.random.randint(1,7, size=1000)
so_many = np.random.randint(1,7, size=10000)
```

## 주사위 결과 빈도수 계산

### 히스토그램

```python
few_count = np.histogram(few, bins=np.arange(.5, 7.5))[0]
many_count = np.histogram(many, bins=np.arange(.5, 7.5))[0]
so_many_count = np.histogram(so_many, bins=np.arange(.5, 7.5))[0]
```

### 주사위 결과 시각화

10번 돌렸을 때 2가 한 번도 안 나왔다고 하면 데이터만 봤을 때 주사위에서 2가 나올 확률은 0  
따라서 적은 샘플로 나온 결과는 주의해야 함.

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(8,3))

ax1.bar(np.arange(1,7), few_count)
ax2.bar(np.arange(1,7), many_count)
ax3.bar(np.arange(1,7), so_many_count)

plt.tight_layout()
plt.show()
```

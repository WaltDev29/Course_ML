# 다변량 선형 회귀 분석

세계보건기구 WHO에서 내어 놓은 각 나라별 기대수명 데이터

## 1. 패키지 설정

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

## 2. 데이터 로드

```python
data_loc = 'https://github.com/dknife/ML/raw/main/data/'
life = pd.read_csv(data_loc + 'life_expectancy.csv')
life.head()
```

```python
life = life[['Life expectancy', 'Year', 'Alcohol',
             'Percentage expenditure', 'Total expenditure',
             'Hepatitis B', 'Measles', 'Polio', 'BMI', 'GDP',
             'Thinness 1-19 years', 'Thinness 5-9 years']]
print(life)
```

## 3. 데이터 전처리

### 3-1. 결손값 처리

```python
print(life.shape)
print(life.isnull().sum())
```

```python
life.dropna(inplace = True)
print(life.shape)
```

### 3-2. 상관관계 분석

#### 3-2-1. 상관행렬

```python
sns.set(rc={'figure.figsize':(12,10)}) # 상관행렬 가시
correlation_matrix = life.corr().round(2) # 상관행렬 생성
sns.heatmap(data=correlation_matrix, annot=True)
plt.show()
```

#### 3-2-2. pair Plot

```python
sns.pairplot(life[['Life expectancy', 'Alcohol', 'Percentage expenditure',
                   'Measles', 'Polio', 'BMI', 'GDP', 'Thinness 1-19 years']])
plt.show()
```

### 3-3. 상관관계가 있는 컬럼만 사용하여 X, Y 데이터 분리

앞 절의 조사를 바탕으로 알콜 소비, 보건 예산, 소아마비 접종률, BMI, GDP, 저체중 정보만을 추출하여서 데이터 프레임으로 재구성

```python
X = life[['Alcohol', 'Percentage expenditure', 'Polio',
          'BMI', 'GDP', 'Thinness 1-19 years']]
y = life['Life expectancy']
print(X)
print(y)
```

### 3-4. Train, Test 데이터 분리

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test1 = train_test_split(X, y, test_size = 0.2)
```

## 4. 모델 학습

```python
from sklearn.linear_model import LinearRegression

lin_model = LinearRegression()
lin_model.fit(X_train, y_train)
```

## 5. 모델 테스트

### 5-1. Train 데이터 정답과 예측치 비교

훈련용 데이터에 대한 예측치를 y_hat_train이라고 하고, 이를 정답인y_train과 비교 하여 화면에 그렸을 때,  
기울기 1인 직선에 가까울 수록 예측과 정답이 일치하는 것

```python
y_hat_train = lin_model.predict(X_train)
plt.scatter(y_train, y_hat_train)
xy_range = [40, 100]
plt.plot(xy_range, xy_range)
```

### 5-2. Test 데이터 정답과 예측치 비교

```python
y_hat_test1 = lin_model.predict(X_test)

plt.scatter(y_test1, y_hat_test1) # 검증 데이터와 예측 데이터의 산포도
plt.plot(xy_range, xy_range)    # (40,40)에서 (10,100)으로 직선
                                # - 데이터가 이 직선과 일치할수록 예측이 정확
```

## 6. 모델 평가

MSE로 평가

```python
from sklearn.metrics import mean_squared_error
print('Mean squared error:', mean_squared_error(y_test1, y_hat_test1))
```

## 7. X 데이터를 정규화하여 학습

### 7-1. 정규화

```python
from sklearn.preprocessing import normalize

n_X = normalize(X, axis = 0) # 정규화를 0번 축 기준으로 실시
```

### 7-2. 모델 학습

```python
nXtrain,nXtest,y_train,y_test2 = train_test_split(n_X, y, test_size=0.2)
lin_model = LinearRegression()
lin_model.fit(nXtrain, y_train)
```

### 7-3. 학습 결과 그래프 표시

```python
y_hat_train = lin_model.predict(nXtrain)
y_hat_test2 = lin_model.predict(nXtest)
plt.scatter(y_train, y_hat_train, color='r')
plt.scatter(y_test2, y_hat_test2, color='b')
plt.plot(xy_range, xy_range)
```

### 7-4. 기존 모델과 비교

```python
plt.scatter(y_test1, y_hat_test1, color='r', label="Model 1")
plt.scatter(y_test2, y_hat_test2, color='b', label="Model 2")
plt.legend()
plt.plot(xy_range, xy_range)
```

### 7-5. 모델 평가

```python
from sklearn.metrics import mean_squared_error
print('Mean squared error:', mean_squared_error(y_test2, y_hat_test2))
```

## 8. 데이터 표준화를 통한 분석 성능 개선

### 8-1. 데이터 히스토그램 확인

```python
plt.hist(n_X, bins=5)
```

### 8-2. 데이터 표준화

```python
from sklearn.preprocessing import StandardScaler

sXtrain,sXtest,y_train,y_test3 = train_test_split(X, y, test_size=0.2)

scaler = StandardScaler()             # 표준화 스케일러 생성
s_X = scaler.fit_transform(sXtrain)         # 표준화 스케일러 적용

plt.hist(s_X, bins=5)                 # 표준화 결과의 가시화
```

### 8-3. 모델 학습 및 그래프 표시

```python
lin_model = LinearRegression()
lin_model.fit(sXtrain, y_train)
y_hat_train = lin_model.predict(sXtrain)
y_hat_test3 = lin_model.predict(sXtest)
plt.scatter(y_train, y_hat_train, color='r')
plt.scatter(y_test3, y_hat_test3, color='b')
plt.plot(xy_range, xy_range)
```

### 8-4. 모델별 비교

```python
plt.scatter(y_test1, y_hat_test1, color='r', label="Model 1")
plt.scatter(y_test2, y_hat_test2, color='g', label="Model 2")
plt.scatter(y_test3, y_hat_test3, color='b', label="Model 3")
plt.legend()
plt.plot(xy_range, xy_range)
```

### 8-5. 오차 확인

```python
print('Mean squared error:', mean_squared_error(y_test3, y_hat_test3))
```

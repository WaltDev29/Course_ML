# KNN 분류 데이터 및 결과 분석

## 1. 패키지 설정

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
import numpy as np
import matplotlib.pyplot as plt
```

## 2. 데이터 준비

```python
train_data = np.array([
    [25,25],
    [33,30],
    [38,30],
    [45,35],
    [28,40]
])

train_label = np.array([0,0,1,1,0])
```

```python
test_data = np.array([[30,35]])
```

## 3. 탐색적 데이터 분석

```python
plt.scatter(train_data[:,0], train_data[:,1], c=train_label)

plt.scatter(test_data[:,0], test_data[:,1], c="red", marker='D', s=100)
plt.xlabel("x1")
plt.xlabel("x2")
plt.show()
```

## 4. 피처 스케일링

모든 데이터에 대해 아래와 같이 스케일링  
평균 ≈ 0  
분산 ≈ 1  

```python
scaler_x = StandardScaler()
scaler_x.fit(train_data)
train_data_std = scaler_x.transform(train_data)
print(train_data_std)
```

```python
test_data_std = scaler_x.transform(test_data)
print(test_data_std)
```

```python
plt.scatter(train_data_std[:,0], train_data_std[:,1], c=train_label)
plt.scatter(test_data_std[:,0], test_data_std[:,1], c="red", marker='D', s=100)
plt.xlabel("x1")
plt.xlabel("x2")
plt.show()
```

## 5. 모형화 및 학습

```python
# 모형화
knn = KNeighborsClassifier(n_neighbors=3, metric="euclidean")

# 학습
knn.fit(train_data_std, train_label)
```

## 6. 예측

```python
# 예측
pred = knn.predict(test_data_std)
print(pred)
```

```python
# 클래스별 확률 값 반환
knn.predict_proba(test_data_std)
```

```python
# 인접한 K개의 개체들에 대한 거리와 색인을 반환
dist, index = knn.kneighbors(train_data_std)
print(dist)
print(index)
```
